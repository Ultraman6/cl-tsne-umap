```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib
import sys
sys.path.append("..")
from scripts.utils import get_path
```

```python
# set up matplotlib
plt.style.use("default")
cmap = matplotlib.cm.get_cmap("tab10")
matplotlib.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams["font.family"] = "serif"
plt.rcParams['font.serif'] = "Times New Roman"
```

```python
fig_path = get_path("figures")
```

```python
d = np.linspace(0.01, 5, 500)
```

```python
attr_umap = np.log(1 + d**2)
attr_neg = np.log(2 + d**2)

rep_umap = np.log((1+d**2) / d**2)
rep_neg = np.log((2+d**2) / (1+d**2))
```

```python

plt.figure(figsize=(5,5))
fontsize = 20
plt.plot(d, attr_umap, label="Attr. UMAP", c=cmap(0))
plt.plot(d, attr_neg, label=r"Attr. Neg-$t$-SNE", c=cmap(1))
plt.plot(d, rep_umap, label="Rep. UMAP", c=cmap(0), linestyle="dotted")
plt.plot(d, rep_neg, linestyle="dotted", c=cmap(1), label=r"Rep. Neg-$t$-SNE")
plt.legend(fontsize=fontsize, loc=(0.1,0.55))
plt.ylim(0, 4)
plt.xlim(0,5)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.xticks(fontsize=fontsize)
plt.xlabel("Embedding distance", fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.savefig(os.path.join(fig_path, "attr_rep_UMAP_neg.png"),
            bbox_inches='tight',
            pad_inches=0,
            dpi=200,)
```
